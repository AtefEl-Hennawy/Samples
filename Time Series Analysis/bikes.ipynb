```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
dt = pd.read_csv(r"train_data.csv")
```

```python
# convert to datetime
dt['Date']= pd.to_datetime(dt['Date'], format='%d/%m/%Y')
```

```python
# no need for that column
dt.drop(columns='timestamp', inplace=True)
```

```python
# setting ID as index
dt.set_index('ID', inplace=True)
```

```python
# exercise where we change binary classifications to zeros and ones
dt['Functioning Day'] = dt['Functioning Day'].apply(lambda x : 1 if x == 'Yes' else 0)
```

```python
# drop rows with non functioning days
rows_to_drop =  dt[dt['Functioning Day'] == 0].index
```

```python
dt.drop(rows_to_drop, inplace=True)
```

```python
# no need for that column anymore
dt.drop(columns='Functioning Day', inplace=True)
```

```python
# change holidays days to ones and the rest to zeros
dt['Holiday'] = dt['Holiday'].apply(lambda x : 1 if x == 'Holiday' else 0)
```

```python
# change seasons to columns with zeros and ones
dummy = pd.get_dummies(dt['Seasons'])
dt = pd.concat([dt,dummy] ,axis='columns')
```

```python
# grouping hours where there is an expected increase in the next hour by +ve values and -ve values where a decrease is expected in the next hour
group_labels = [-1, 5, -10, -1, 1, 10, -5, -1]
bins = [0,4,8,9,10,17,18,19,24]
dt['increasing hrs'] = pd.cut(dt['Hour'],bins=bins,labels=group_labels, right=False, duplicates='drop', ordered=False, include_lowest=True)
```

```python
dt['increasing hrs'] = dt['increasing hrs'].astype(int)
```

```python
# adding a day column
dt['Day'] = dt['Date'].dt.day_name()
```

```python
# hours rental rate
dt[['y', 'Day']].groupby('Day').sum('y').plot.bar(figsize=(20, 5));
dt.plot.scatter('Day', 'y', figsize=(20, 5));
```

```python
# mean of rented bikes by day, notice that we need to predict those empty days
dt.resample('D', on='Date')['y'].mean().plot(figsize=(20, 5));
```

```python
# sample month of above figure
dt.resample('D', on='Date')['y'].mean().loc['2018-04-01':'2018-04-30'].plot(figsize=(20, 5));
```

```python
# making a time series column with both date and hours 
# a bit of a messy code by the idea is to get the date as a string from date column then making a time series of the hour column to  
# convert 0, 1, 5 hours to 00, 01, 05 hours then get as a string 
# concatane those strings to in a single column then convert it to time series 

dt['hours2'] = dt['Date'].dt.strftime("%Y-%m-%d ") + pd.to_datetime(dt['Hour'], format='%H').dt.strftime("%H")
dt['hours2'] = pd.to_datetime(dt['hours2'], format='%Y-%m-%d %H')
```

```python
# mean of rented bikes by hour, notice that we need to predict those empty hours
dt.resample('H', on='hours2')['y'].mean().plot(figsize=(20, 5));
```

```python
# sample day of above figure
dt.resample('H', on='hours2')['y'].sum().loc['2018-06-18 00:00:00':'2018-06-19 00:00:00'].plot(figsize=(20, 5));
```
